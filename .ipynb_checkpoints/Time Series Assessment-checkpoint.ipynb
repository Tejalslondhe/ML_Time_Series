# Q.2) Perform time series analysis:
#### a) Visualize time series data

```python
for i in data.columns:
    data[i].plot(figsize=(15,3))
    plt.show()
```

#### b) Check Stationarity with :
####  ADF Test

* Ho : - The data is  non-stationary
* H1 : - The data is  stationary

* Alpha = 0.05

```python
from statsmodels.tsa.stattools import adfuller

for i in data.columns:
    datatest = adfuller(data[i])
    dataoutput=pd.Series(datatest[0:4],index=['Test Statistic','p-value','#lags used','number of observations used'])
    print('\nADF test for',i,'\n',dataoutput)
    if dataoutput['p-value'] > 0.05:
        print('We do not reject Ho\n','Data is Non-Stationary')
    else:
        print('We reject Ho\n','Data is Stationary')
```

#### KPSS Test

* Ho : - The data is  Stationary
* H1 : - The data is  Non-Stationary

* Alpha = 0.05

```python
from statsmodels.tsa.stattools import kpss

for i in data.columns:
    datatest = kpss(data[i])
    dataoutput=pd.Series(datatest[0:4],index=['Test Statistic','p-value','#lags used','Critical values'])
    print('\nKPSS test for',i,'\n',dataoutput)
    if dataoutput['p-value'] > 0.05:
        print('We do not reject Ho\n','Data is Stationary')
    else:
        print('We reject Ho\n','Data is Non-Stationary')
```
